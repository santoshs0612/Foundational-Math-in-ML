# Implimenting the basic Gradient decent


```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
np.random.seed(42)
# 50 evenly spaced points between 0 and 10
X = np.linspace(0,10,50)
true_m , true_c = 2 ,1
y = true_m * X + true_c + np.random.normal(0,2,X.shape)
```

```python
plt.figure(figsize=(10,6))
plt.scatter(X,y,label = " Data Points", color="green")
plt.title("Input Data")
plt.xlabel("X")
plt.ylabel("y")
plt.legend()
plt.show()
```

```python
m , c  = 0 , 0
learning_rate = 0.001
epochs = 1000
```

```python
#Store losses
losses = []
```

# Gradient calculation

```python
for i in range(epochs):
  #prediced y values
  y_pred = m * X + c

  #compute gradient
  m_grad = (-2/X.shape[0]) * np.sum(X*(y-y_pred))
  c_gred = (-2/X.shape[0]) * np.sum(y-y_pred)

  #Update the parameters
  m = m - learning_rate * m_grad
  c = c - learning_rate * c_gred
  #Compute loss(MSE)
  loss = np.mean((y-y_pred)**2)
  loss_history = losses.append(loss)
  # print updates every 10 epoches
  if i % 100==0:
    print(f"Iteration {i}: Loss = {loss}, m = {m:.4f},c = {c:.4f} ")
```

```python
plt.figure(figsize=(10,6))
plt.scatter(X,y,label = " Data Points", color="green")
plt.plot(X,y_pred,label = "Predicted Line", color="red")
plt.title("Gradinet Decent:")
plt.xlabel("X")
plt.ylabel("y")
plt.legend()
plt.show()
```
