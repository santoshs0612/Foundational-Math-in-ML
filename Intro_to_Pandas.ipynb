### Pandas Project

```python
url = "https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv"
import pandas as pd
df = pd.read_csv(url)
print(df.head())
```

```python
# df.info()
df.describe()
```

```python
df.isnull().sum()
```

```python
df['Age'].fillna(df['Age'].mean(),inplace=True)
df.info()
df.drop(['Cabin'],axis=1,inplace=True)
df.info()
```

# Analyze survival Pattern
  - What is the overall survival rate?
  - What is the survival rate by gender?
  - What is the survival rate by passenger class(Pclasss)?
  - What is the average age of survival vs non-survivals?
  - What is the survival rate of children (Age<16) compared to adults?

```python
df['Survived'].mean()*100
```

```python
df.groupby('Sex')['Survived'].mean()*100
```

```python
df.groupby('Pclass')['Survived'].mean()*100
```

```python
df.groupby('Survived')['Age'].mean()
```

```python
df[(df['Age']>16)]['Survived'].mean()*100
df[(df['Age']<16)]['Survived'].mean()*100
```

```python
young = df[df['Survived']==1].sort_values(by='Age',ascending=True)
print(young['Name'])
```

```python
bins = [0,16,60,float('inf')]
labels = ['Children','Adult','Senior']
df['Age_Group'] = pd.cut(df['Age'],bins=bins,labels=labels)
df.groupby('Age_Group')['Survived'].mean()*100
```
